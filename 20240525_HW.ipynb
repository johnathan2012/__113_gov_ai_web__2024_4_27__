```python
from pydantic import BaseModel, Field

class FactoryInfo(BaseModel):
    fname:str = Field(alias="Name_Ins")
    reg_no:str = Field(alias="No")
    addr:str = Field(alias="Address")
    tax_ID_number:str
    lat:str = Field(alias="wgs84aX")
    lon:str = Field(alias="wgs84aY")

class Result(BaseModel):
    records:list[FactoryInfo]

class Factories(BaseModel):
    result:Result

import json
with open(file="新北市食品工廠清冊.json",encoding="utf-8") as file:
    content:dict = json.load(file)

    Finfo = Factories.model_validate(content)
    factories:list[Factories] = Finfo.result.records

for f_info in factories:
    print(f_info)
```

```python
from pydantic import BaseModel,Field
import json

class Fac_detail(BaseModel):
    name:str = Field(alias="Name_Ins")
    no:str = Field(alias="No")
    addr:str = Field(alias="Address")
    tax_ID_number:str
    cord_X:str = Field(alias="wgs84aX")
    cord_Y:str = Field(alias="wgs84aY")

class Result(BaseModel):
    records:list[Fac_detail] 

class Fac_list(BaseModel):
    result:Result

with open(file="新北市食品工廠清冊.json",encoding="utf-8") as file:
    content:dict = json.load(file)
    print(content)
    data_list = Fac_list.model_validate(content)


fac_list:list[Fac_list] = data_list.result.records

print(f"共取出{len(fac_list)}筆資料")

for each_fac in fac_list:
    print(each_fac)
```
